```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
#fins aquí entrem a sakila
rentals_may= pd.read_sql_query('SELECT * FROM rentals_may', engine) #de la mateixa manera que llegim csv, llegim sql_query
rentals_may.head()
```

```python
rentals_june= pd.read_sql_query('SELECT * FROM rentals_june', engine)
rentals_june.head()
```

```python
num_customer_rentals_may = pd.read_sql_query('select count(rental_id) as num_rent_may, customer_id from rentals_may group by customer_id', engine)
num_customer_rentals_may
```

```python
num_customer_rentals_june = pd.read_sql_query('select count(rental_id) as num_rent_june, customer_id from rentals_june group by customer_id', engine)
num_customer_rentals_june
```

```python
comparison = num_customer_rentals_may.merge(num_customer_rentals_june, on='customer_id', how='inner')
comparison
```

```python
lmb = lambda x: 'june' if x['num_rent_june'] > x['num_rent_may'] else 'may'
Comparison.apply(lmb, axis = 1)
```

```python
comparison['most_rents']=Comparison.apply(lmb, axis = 1)
comparison
```
